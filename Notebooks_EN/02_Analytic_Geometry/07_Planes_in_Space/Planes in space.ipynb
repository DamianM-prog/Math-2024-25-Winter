# Analytic Geometry: Planes in Space

In analytic geometry, a **plane in space** is a two-dimensional flat surface that extends infinitely in all directions. It can be described using several methods, most commonly using **equations**. The most basic way to represent a plane in space is with a linear equation in three variables.

## Plane Equation in General Form

The general equation of a plane in 3D space is:

$$
Ax + By + Cz + D = 0
$$

Where:
- \( A \), \( B \), and \( C \) are the coefficients that define the **normal vector** of the plane.
- \( D \) is a constant that shifts the plane along the axes.
- \( x \), \( y \), and \( z \) are the coordinates of a point on the plane.

The **normal vector** \($ \vec{n} = (A, B, C) $\) is a vector that is perpendicular to the plane.

### Example:
Consider the equation of the plane:

$$
2x - 3y + 4z = 5
$$

Here, the normal vector is \($ \vec{n} = (2, -3, 4) $\).

## Point-Normal Form of the Plane

Another way to express the equation of a plane is in **point-normal form**. If we know a point \($ P_0(x_0, y_0, z_0) $\) on the plane and the normal vector \( $\vec{n} = (A, B, C) $\), the equation of the plane can be written as:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

Where \( (x_0, y_0, z_0) \) is a known point on the plane.

### Example:
Given the normal vector \($ \vec{n} = (2, -3, 4) $\) and the point \($ P_0(1, 2, 3) $\), the equation of the plane is:

$$
2(x - 1) - 3(y - 2) + 4(z - 3) = 0
$$

Simplifying:

$$
2x - 3y + 4z = 20
$$

## Parametric Equation of the Plane

A **parametric equation** of a plane expresses each point on the plane as a linear combination of two independent direction vectors and a fixed point on the plane.

Let \($ \vec{r}(t, s) $\) represent a point on the plane, where \( t \) and \( s \) are parameters. If \($ \vec{r_0} = (x_0, y_0, z_0) $\) is a point on the plane, and \($ \vec{v_1} = (a_1, b_1, c_1) $\) and \($ \vec{v_2} = (a_2, b_2, c_2) $\) are two non-parallel vectors lying in the plane, the parametric equations are:

$$
\vec{r}(t, s) = \vec{r_0} + t \cdot \vec{v_1} + s \cdot \vec{v_2}
$$

Where:
- \($ \vec{r_0} $\) is a fixed point on the plane.
- \($ \vec{v_1} $\) and \($ \vec{v_2} $\) are direction vectors parallel to the plane.

### Example:
Given \($ \vec{r_0} = (1, 2, 3) $\), \($ \vec{v_1} = (1, 0, 0) $\), and \($ \vec{v_2} = (0, 1, 0) $\), the parametric equations of the plane are:

$$
\vec{r}(t, s) = (1, 2, 3) + t \cdot (1, 0, 0) + s \cdot (0, 1, 0)
$$

Or:

$$
\vec{r}(t, s) = (1 + t, 2 + s, 3)
$$

## How to Find the Equation of a Plane

To find the equation of a plane, you need:
1. A point on the plane.
2. A normal vector to the plane (or two non-parallel direction vectors).

### Example:

Given two points on the plane \($ P_1(1, 2, 3) $\) and \($ P_2(4, 5, 6) $\), and the normal vector \($ \vec{n} = (2, -1, 3) $\), the equation of the plane is:

1. Take the point \($ P_1 $\) and the normal vector \($ \vec{n} $\).
2. Plug into the point-normal form:

$$
2(x - 1) - 1(y - 2) + 3(z - 3) = 0
$$

Simplifying:

$$
2x - y + 3z = 11
$$

Thus, the equation of the plane is \($ 2x - y + 3z = 11 $\).

$$

$$

$$
\mathbf{AC} = \mathbf{C} - \mathbf{A} = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)
$$

Next, we calculate the cross product \($ \mathbf{AB} \times \mathbf{AC} $\) using the determinant of a matrix:

$$
\mathbf{AB} \times \mathbf{AC} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 2 & 2 \\ 1 & -1 & 1 \end{vmatrix}
$$

We compute each component of the cross product:

- For \($ \hat{i} $\):
  $$
  \hat{i} \left( \begin{vmatrix} 2 & 2 \\ -1 & 1 \end{vmatrix} \right) = \hat{i} \left( (2)(1) - (2)(-1) \right) = \hat{i} (2 + 2) = 4\hat{i}
  $$

- For \($ \hat{j}$ \):
  $$
  -\hat{j} \left( \begin{vmatrix} 2 & 2 \\ 1 & 1 \end{vmatrix} \right) = -\hat{j} \left( (2)(1) - (2)(1) \right) = -\hat{j} (2 - 2) = 0\hat{j}
  $$

- For \($ \hat{k} $\):
  $$
  \hat{k} \left( \begin{vmatrix} 2 & 2 \\ 1 & -1 \end{vmatrix} \right) = \hat{k} \left( (2)(-1) - (2)(1) \right) = \hat{k} (-2 - 2) = -4\hat{k}
  $$

Thus, the cross product is:

$$
\mathbf{AB} \times \mathbf{AC} = 4\hat{i} - 0\hat{j} - 4\hat{k} = (4, 0, -4)
$$


Next, we use the point \( A(1, 2, 3) \) and the cross product \($ \mathbf{AB} \times \mathbf{AC} = (4, 0, -4) $\) to write the equation of the plane:

$$

Simplifying:

$$
4x - 4 - 4z + 12 = 0
$$

$$

$$

Finally, dividing the equation by 4:

$$

The direction vector of the line of intersection is the cross product of the normal vectors of the two planes.

- The normal vector to Plane 1, $ \mathbf{n_1} $, is \( [2, 3, 4] \).
- The normal vector to Plane 2, $ \mathbf{n_2} $, is \( [3, 4, 2] \).

To compute the cross product \($ \mathbf{d} = \mathbf{n_1} \times \mathbf{n_2} $\), use the formula:

$$
\mathbf{d} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 3 & 4 \\ 3 & 4 & 2 \end{vmatrix}
$$

Compute each component:

- $ \hat{i}: \begin{vmatrix} 3 & 4 \\ 4 & 2 \end{vmatrix} = (3)(2) - (4)(4) = 6 - 16 = -10 $
- $ \hat{j}: \begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} = (2)(2) - (4)(3) = 4 - 12 = -8 $
- $ \hat{k}: \begin{vmatrix} 2 & 3 \\ 3 & 4 \end{vmatrix} = (2)(4) - (3)(3) = 8 - 9 = -1 $

Thus, the cross product gives the direction vector:

$$
\mathbf{d} = (-10, 8, -1)
$$

### Now find a point on the line of intersection.

To find a point on the line, we solve the system of two plane equations. We can do this by setting one of the variables to a specific value and solving for the other two.

Let's set \( z = 0 \) (for simplicity) and substitute into the two plane equations.

- Plane 1: \($ 2x + 3y + 4(0) = 5 \quad \Rightarrow \quad 2x + 3y = 5 $\)
- Plane 2: \($ 3x + 4y + 2(0) = 6 \quad \Rightarrow \quad 3x + 4y = 6 $\)

Now we solve this system of equations:

1. \( 2x + 3y = 5 \)
2. \( 3x + 4y = 6 \)

To solve, multiply the first equation by 3 and the second equation by 2 to eliminate \( x \):

$$
6x + 9y = 15 \quad \text{(Multiply equation 1 by 3)}
$$

$$
6x + 8y = 12 \quad \text{(Multiply equation 2 by 2)}
$$

Now subtract the second equation from the first:

$$
(6x + 9y) - (6x + 8y) = 15 - 12
$$

$$

$$
2x + 3(3) = 5 \quad \Rightarrow \quad 2x + 9 = 5 \quad \Rightarrow \quad 2x = -4 \quad \Rightarrow \quad x = -2

Thus, a point on the line of intersection is \( (-2, 3, 0) \).

- \( A = 2 \), \( B = 3 \), \( C = 4 \) are the coefficients from the plane equation.

$$
