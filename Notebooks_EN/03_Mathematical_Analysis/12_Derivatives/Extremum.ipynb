# Extremum (Maximum and Minimum)

In mathematics, the **extremum** refers to the points at which a function reaches either its maximum or minimum value. These points are critical in optimization problems, where we aim to either maximize or minimize a given quantity.

## Types of Extremum

1. **Local Maximum**: A point \($ x = c $\) is a local maximum of a function \($ f(x) $\) if \($ f(c) \geq f(x) $\) for all \($ x $\) in some neighborhood around \($ c $\). In other words, \($ f(c) $\) is larger than or equal to the values of \($ f(x) $\) near \($ c $\).

2. **Local Minimum**: A point \($ x = c $\) is a local minimum of a function \($ f(x) $\) if \($ f(c) \leq f(x) $\) for all \($ x $\) in some neighborhood around \($ c $\). In other words, \($ f(c) $\) is smaller than or equal to the values of \($ f(x) $\) near \($ c $\).

3. **Global Maximum**: A point \($ x = c $\) is a global maximum of a function \($ f(x) $\) if \($ f(c) \geq f(x) $\) for all values of \($ x $\) in the domain of the function. The value \($ f(c) $\) is the largest value of the function across its entire domain.

4. **Global Minimum**: A point \($ x = c $\) is a global minimum of a function \($ f(x) $\) if \($ f(c) \leq f(x) $\) for all values of \($ x $\) in the domain of the function. The value \($ f(c) $\) is the smallest value of the function across its entire domain.

## Finding Extrema

To find the extrema (both local and global) of a function \($ f(x) $\), follow these steps:

1. **Find the first derivative**: Take the derivative of \($ f(x) $\) to find \($ f'(x) $\). Critical points occur where the first derivative is zero or undefined.

2. **Solve \($ f'(x) = 0 $\)**: Set the first derivative equal to zero and solve for \($ x $\). These are the critical points, where the function may have a local extremum.

3. **Use the second derivative test (optional)**:
   - If \($ f''(x) > 0 $\) at a critical point \($ x = c $\), then \($ f(c) $\) is a local minimum.
   - If \($ f''(x) < 0 $\) at a critical point \($ x = c $\), then \($ f(c) $\) is a local maximum.
   - If \($ f''(x) = 0 $\), the second derivative test is inconclusive, and you may need to use other methods (such as the first derivative test).

4. **Check endpoints (if applicable)**: For a function with a restricted domain, check the function's values at the endpoints of the domain.

## Example

For the function \($ f(x) = x^2 + 3x - 5 $\), we want to find the extrema:

1. First, find the derivative of \($ f(x) $\):

   \[$
   f'(x) = 2x + 3
   $\]

2. Set the derivative equal to zero to find the critical points:

   \[$
   2x + 3 = 0 \quad \Rightarrow \quad x = -\frac{3}{2}
   $\]

3. Use the second derivative test:

   \[
   f''(x) = 2
   \]

   Since \($ f''(x) > 0 $\), the critical point \($ x = -\frac{3}{2} $\) is a local minimum.

## Conclusion

An **extremum** can either be a **maximum** or **minimum** depending on whether the function reaches a peak or a trough at that point. The process of finding extrema involves differentiating the function, solving for critical points, and using tests to classify those points as maxima or minima.

Given the function:

$$
f(x) = \frac{x^2 + 2x + 1}{x - 1}
$$

This can be simplified as:

$$
f(x) = \frac{(x + 1)^2}{x - 1}
$$

We know that \($ f(x) = \frac{g(x)}{h(x)} $\), where:

$$
g(x) = (x + 1)^2
$$
$$
h(x) = x - 1
$$

Using the quotient rule for derivatives:

$$
f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{[h(x)]^2}
$$

Now, compute the derivatives of \($ g(x) $\) and \($ h(x) $\):
- \($ g'(x) = 2(x + 1) $\)
- \($ h'(x) = 1 $\)

Thus:

$$
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2 \cdot 1}{(x - 1)^2}
$$

Simplifying:

$$
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2}{(x - 1)^2}
$$
$$
= \frac{2(x^2 - 1) - (x^2 + 2x + 1)}{(x - 1)^2}
$$
$$
= \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2}
$$
$$
= \frac{x^2 - 2x - 3}{(x - 1)^2}
$$

To find the critical points, set the numerator equal to zero:

$$
\frac{x^2 - 2x - 3}{(x - 1)^2} = 0
$$

Thus, the numerator must be equal to zero:

$$
x^2 - 2x - 3 = 0
$$

Factoring the quadratic equation:

$$
x^2 - 2x - 3 = (x - 3)(x + 1) = 0
$$

So, \($ x = 3 $\) or \($ x = -1 $\).

$$

$$

Since \($ f'(x) $\) changes from negative to positive as \($ x $\) increases through \($ x = 3 $\), \($ x = 3 $\) is a **local minimum**.

#### 2. At \($ x = -1 $\):
