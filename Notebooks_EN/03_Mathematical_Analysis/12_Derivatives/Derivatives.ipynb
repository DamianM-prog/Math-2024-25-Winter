# Introduction to Derivatives

## What is a Derivative?

A **derivative** represents the rate of change of a function with respect to one of its variables. In simpler terms, it measures how the output of a function changes as the input changes. The derivative of a function can be thought of as the slope of the function's graph at any given point.

Mathematically, the derivative of a function \( f(x) \) at a point \( x \) is denoted as:

\[$
f'(x) = \frac{d}{dx} f(x)
$\]

The derivative of a function tells us how sensitive the function's value is to changes in \( x \). If the derivative is large, small changes in \( x \) lead to large changes in \( y \). If it's small, small changes in \( x \) lead to small changes in \( y \).

## Basic Derivative Rules

### 1. **Constant Rule**
If \($ f(x) = c $\), where \( c \) is a constant, then the derivative is:

\[$
f'(x) = 0
$\]

### 2. **Power Rule**
If \($ f(x) = x^n $\), where \( n \) is any real number, then:

\[$
f'(x) = n \cdot x^{n-1}
$\]

### 3. **Sum Rule**
If \($ f(x) = g(x) + h(x) $\), then:

\[$
f'(x) = g'(x) + h'(x)
$\]

### 4. **Product Rule**
If \($ f(x) = g(x) \cdot h(x) $\), then:

\[$
f'(x) = g'(x) \cdot h(x) + g(x) \cdot h'(x)
$\]

### 5. **Quotient Rule**
If \($ f(x) = \frac{g(x)}{h(x)} $\), then:

\[$
f'(x) = \frac{g'(x) \cdot h(x) - g(x) \cdot h'(x)}{h(x)^2}
$\]

### 6. **Chain Rule**
If \($ f(x) = g(h(x)) $\), then the derivative is:

\[$
f'(x) = g'(h(x)) \cdot h'(x)
$\]

## Geometric Interpretation

The derivative of a function at a specific point can be interpreted as the **slope of the tangent line** to the graph of the function at that point. If the derivative is positive, the function is increasing; if the derivative is negative, the function is decreasing.

## Common Derivatives

Here are some common derivatives for quick reference:

- **\($ \frac{d}{dx} (x^n) = n \cdot x^{n-1} $\)**
- **\($ \frac{d}{dx} (\sin(x)) = \cos(x) $\)**
- **\($ \frac{d}{dx} (\cos(x)) = -\sin(x) $\)**
- **\($ \frac{d}{dx} (e^x) = e^x $\)**
- **\($ \frac{d}{dx} (\ln(x)) = \frac{1}{x} $\)**

## Applications of Derivatives

Derivatives are used in various fields such as physics, economics, biology, and engineering. Some common applications include:

- **Finding the velocity** of an object in motion (rate of change of position).
- **Optimization problems** (finding the maximum or minimum of a function).
- **Graph analysis**, such as determining intervals of increase or decrease and concavity of functions.

\[$
\frac{d}{dx} \left( \ln(u) \right) = \frac{1}{u} \cdot \frac{du}{dx}

\[$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \frac{1}{\sin(x)} \cdot \cos(x)
$\]

\[$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \frac{\cos(x)}{\sin(x)}

f'(x) = \frac{d}{dx} \left( \cos(x) \right)

\[$
f'(x) = -\sin(x)

\[$
f''(x) = \frac{d}{dx} \left( -\sin(x) \right)
$\]

5. The derivative of \($ -\sin(x) $\) is \($ -\cos(x) $\), so:

\[$
f''(x) = -\cos(x)
$\]

\[$
f''(x) = -\cos(x)
$\]

# L'Hopital's Rule - Improper Limits

### 1. Limit 1: \($ \lim_{x \to 0} \frac{\sin(x)}{x} $\)

- The derivative of \($ \sin(x) $\) is \($ \cos(x) $\).
- The derivative of \($ x $\) is 1.

Applying L'Hopital's Rule:

\[$
\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} = \cos(0) = 1
$\]

### 2. Limit 2: \($ \lim_{x \to \infty} \frac{\ln(x)}{x} $\)

- The derivative of \($ \ln(x) $\) is \($ \frac{1}{x} $\).
- The derivative of \($ x $\) is 1.

Applying L'Hopital's Rule:

\[$
\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \lim_{x \to \infty} \frac{1}{x} = 0
$\]

### 3. Limit 3: \($ \lim_{x \to \infty} \frac{e^x}{x} $\)

- The derivative of \($ e^x $\) is \($ e^x $\).
- The derivative of \($ x $\) is 1.

Applying L'Hopital's Rule:

\[$
\lim_{x \to \infty} \frac{e^x}{x} = \lim_{x \to \infty} \frac{e^x}{1} = \lim_{x \to \infty} e^x = \infty
$\]
