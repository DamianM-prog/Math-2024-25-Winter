# Calculations and Limits

### 1. **Limit of the Rational Function**

We need to calculate the following limit:

\[$
\lim_{n \to \infty} \frac{n^2 + 3n}{n^2 - 2n}
$\]

To solve this limit, divide both the numerator and denominator by \($n^2$\) (the highest power of \(n\)):

\[$
\frac{n^2 + 3n}{n^2 - 2n} = \frac{1 + \frac{3}{n}}{1 - \frac{2}{n}}
$\]

Now, take the limit as \($n \to \infty$\):

\[$
\lim_{n \to \infty} \frac{1 + \frac{3}{n}}{1 - \frac{2}{n}} = \frac{1 + 0}{1 - 0} = 1
$\]

Thus, the limit is:

\[$
\lim_{n \to \infty} \frac{n^2 + 3n}{n^2 - 2n} = 1
$\]

### 2. **Limit of the Rational Function Involving Cubes**

We need to calculate the following limit:

\[$
\lim_{n \to \infty} \frac{(2n + 3)}{3n^3 - 1}
$\]

First, note that for large \(n\), the highest power of \(n\) dominates the behavior of both the numerator and the denominator. The numerator is linear (\($2n + 3$\)), and the denominator is cubic (\($3n^3 - 1$\)).

The degree of the denominator (3) is greater than the degree of the numerator (1), which suggests that the limit will approach 0 as \(n\) approaches infinity.

We can verify this rigorously by dividing both the numerator and denominator by \($n^3$\):

\[$
\frac{(2n + 3)}{3n^3 - 1} = \frac{\frac{2}{n^2} + \frac{3}{n^3}}{3 - \frac{1}{n^3}}
$\]

Now, take the limit as \($n \to \infty$\):

\[$
\lim_{n \to \infty} \frac{\frac{2}{n^2} + \frac{3}{n^3}}{3 - \frac{1}{n^3}} = \frac{0 + 0}{3 - 0} = 0
$\]

Thus, the limit is:

\[$
\lim_{n \to \infty} \frac{(2n + 3)}{3n^3 - 1} = 0
$\]

### 3. **Prove Using the Squeeze Theorem**

We are asked to prove using the Squeeze Theorem that:

\[$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$\]

Since \($\sin(n)$\) is always bounded between -1 and 1 for all \($n$\), we have the following inequality:

\[$
-1 \leq \sin(n) \leq 1
$\]

Dividing the entire inequality by \(n\), we get:

\[$
\frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$\]

As \($n \to \infty$\), both \($\frac{-1}{n}$\) and \($\frac{1}{n}$\) approach 0. By the Squeeze Theorem:

\[$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$\]

### 4. **Limit of the Sequence**

We are asked to find the limit of the sequence:

\[$
a_n = \left( 1 + \frac{1}{n} \right)^n
$\]

This is a standard limit, and it is known that:

\[$
\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e
$\]

Thus, the limit is:

\[$
\lim_{n \to \infty} a_n = e
$\]

We are given the following inequality for \($ \sin(n) $\):

$$
-1 \leq \sin(n) \leq 1
$$

Dividing all terms by \(n\), we get:

$$
-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

Now, apply the **Squeeze Theorem**. We know that:

$$
\lim_{n \to \infty} -\frac{1}{n} = 0 \quad \text{and} \quad \lim_{n \to \infty} \frac{1}{n} = 0
$$

Since \($ \frac{\sin(n)}{n} $\) is squeezed between \($ -\frac{1}{n} $\) and \($ \frac{1}{n} $\), we conclude that:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$
