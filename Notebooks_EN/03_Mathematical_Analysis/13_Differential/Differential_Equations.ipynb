# Differential Equations

A **differential equation** is an equation that relates a function to its derivatives. It describes how a quantity changes over time or space. Differential equations can be classified into several types, such as ordinary differential equations (ODEs) and partial differential equations (PDEs).

## 1. Types of Differential Equations

### Ordinary Differential Equations (ODEs)
An ODE involves functions of a single independent variable and its derivatives. The general form is:

$$ F\left(x, y, \frac{dy}{dx}, \frac{d^2y}{dx^2}, \dots \right) = 0 $$

Example:  
$$ \frac{dy}{dx} = y $$

### Partial Differential Equations (PDEs)
A PDE involves functions of multiple independent variables and their partial derivatives. The general form is:

$$ F\left(x, y, z, \frac{\partial y}{\partial x}, \frac{\partial y}{\partial t}, \dots \right) = 0 $$

Example:  
$$ \frac{\partial^2 u}{\partial t^2} - c^2 \frac{\partial^2 u}{\partial x^2} = 0 $$  
This is the **wave equation**.

---

## 2. Solving Ordinary Differential Equations (ODEs)

### First-Order ODEs

A first-order ODE is of the form:

$$ \frac{dy}{dx} = f(x, y) $$

#### Separable Equations
Separable equations can be written as:

$$ \frac{dy}{dx} = g(x) h(y) $$

To solve, separate the variables and integrate:

$$ \frac{1}{h(y)} \, dy = g(x) \, dx $$

### Linear ODEs

A first-order linear ODE has the form:

$$ \frac{dy}{dx} + P(x) y = Q(x) $$

The solution involves finding an integrating factor \( \mu(x) \):

$$ \mu(x) = e^{\int P(x) \, dx} $$

Multiply through by \( \mu(x) \) and then integrate.

### Exact Equations

An ODE is exact if it can be written as:

$$ M(x, y) \, dx + N(x, y) \, dy = 0 $$

where \( \frac{\partial M}{\partial y} = \frac{\partial N}{\partial x} \).

The solution is obtained by finding a potential function \( \Phi(x, y) \) such that:

$$ \frac{\partial \Phi}{\partial x} = M(x, y) \quad \text{and} \quad \frac{\partial \Phi}{\partial y} = N(x, y) $$

---

## 3. Higher-Order ODEs

A higher-order ODE involves derivatives of order higher than one. The general form is:

$$ F\left(x, y, y', y'', \dots, y^{(n)}\right) = 0 $$

### Second-Order Linear ODEs

The standard form of a second-order linear ODE is:

$$ y'' + p(x) y' + q(x) y = r(x) $$

The solution involves finding the complementary function (solution to the homogeneous equation) and a particular solution (for the non-homogeneous equation).

### Homogeneous vs Non-Homogeneous Equations

- **Homogeneous:** If \($ r(x) = 0 $\), the equation is homogeneous.
- **Non-Homogeneous:** If \($ r(x) \neq 0 $\), the equation is non-homogeneous.

---

## 4. Methods of Solving ODEs

### Method of Separation of Variables

Used when the equation is separable. For example:

$$ \frac{dy}{dx} = y \quad \Rightarrow \quad \frac{dy}{y} = dx $$

Integrate both sides:

$$ \int \frac{1}{y} \, dy = \int dx $$

### The Integrating Factor Method

For first-order linear ODEs:

$$ \frac{dy}{dx} + P(x) y = Q(x) $$

Multiply through by the integrating factor \($ \mu(x) = e^{\int P(x) \, dx} $\), then solve.

### Characteristic Equation (for Linear ODEs)

For second-order linear ODEs with constant coefficients:

$$ y'' + a_1 y' + a_2 y = 0 $$

The characteristic equation is:

$$ r^2 + a_1 r + a_2 = 0 $$

Solve for \( r \), and the general solution depends on the nature of the roots.

---

## 5. Second-Order Linear ODEs with Constant Coefficients

Consider the equation:

$$ y'' + p y' + q y = 0 $$

The solution depends on the discriminant of the characteristic equation \($ r^2 + p r + q = 0 $\):

- **Real and distinct roots:** \($ y(x) = C_1 e^{r_1 x} + C_2 e^{r_2 x} $\)
- **Real and repeated roots:** \($ y(x) = (C_1 + C_2 x) e^{r x} $\)
- **Complex roots:** \($ y(x) = e^{\alpha x} (C_1 \cos(\beta x) + C_2 \sin(\beta x)) $\), where \($ \alpha \pm i \beta $\) are the roots.

---

## 6. Applications of Differential Equations

Differential equations have numerous applications, including:

- **Physics**: Modeling motion, heat, waves, etc.
- **Biology**: Population dynamics, spread of diseases.
- **Economics**: Modeling growth rates, economic systems.
- **Engineering**: Electrical circuits, mechanical vibrations.

---

## 7. Summary

- Differential equations describe how quantities change.
- They can be **ordinary** (involving a single variable) or **partial** (involving multiple variables).
- Methods of solving include **separation of variables**, **integrating factors**, and **characteristic equations**.
- Second-order linear equations have distinct solution forms based on the roots of the characteristic equation.

## \($ \frac{dy}{dx} = y $\)

1. **Rearrange the equation** to separate the variables:
   $$ \frac{dy}{dx} = y \quad \Rightarrow \quad \frac{dy}{y} = dx $$

2. **Integrate both sides**:
   $$ \int \frac{1}{y} \, dy = \int 1 \, dx $$

   This gives:
   $$ \ln |y| = x + C $$

3. **Exponentiate both sides** to solve for \($y$\):
   $$ |y| = e^{x + C} = e^C \cdot e^x $$

   Let \($ A = e^C $\), so we get:
   $$ y = A e^x $$

The solution to the first equation is:
$$ y = A e^x $$


##  \($ \frac{dy}{dx} = \frac{1}{2} y $\)

1. **Rearrange the equation**:
   $$ \frac{dy}{dx} = \frac{1}{2} y \quad \Rightarrow \quad \frac{dy}{y} = \frac{1}{2} dx $$

2. **Integrate both sides**:
   $$ \int \frac{1}{y} \, dy = \int \frac{1}{2} \, dx $$

   This gives:
   $$ \ln |y| = \frac{x}{2} + C $$

3. **Exponentiate both sides** to solve for \($y$\):
   $$ |y| = e^{\frac{x}{2} + C} = e^C \cdot e^{\frac{x}{2}} $$

   Let \($ A = e^C $\), so we get:
   $$ y = A e^{\frac{x}{2}} $$

The solution to the second equation is:
$$ y = A e^{\frac{x}{2}} $$

## Exercise 2 \($ \frac{dy}{dx} = \frac{x}{y} $\)

1. **Rearrange the equation**:
   $$ \frac{dy}{dx} = \frac{x}{y} \quad \Rightarrow \quad y \, dy = x \, dx $$

2. **Integrate both sides**:
   $$ \int y \, dy = \int x \, dx $$

   This gives:
   $$ \frac{y^2}{2} = \frac{x^2}{2} + C $$

3. **Solve for \(y\)**:
   $$ y^2 = x^2 + 2C $$

   Taking the square root of both sides:
   $$ y = \pm \sqrt{x^2 + 2C} $$

Thus, the solution to the third equation is:
$$ y = \pm \sqrt{x^2 + 2C} $$

## \($ \frac{dy}{dx} = \frac{y}{x} $\)


1. **Rearrange the equation**:
   $$ \frac{dy}{dx} = \frac{y}{x} \quad \Rightarrow \quad \frac{dy}{y} = \frac{dx}{x} $$

2. **Integrate both sides**:
   $$ \int \frac{1}{y} \, dy = \int \frac{1}{x} \, dx $$

   This gives:
   $$ \ln |y| = \ln |x| + C $$

3. **Exponentiate both sides** to solve for \($y$\):
   $$ |y| = e^{\ln |x| + C} = e^C \cdot e^{\ln |x|} $$

   Since \($ e^{\ln |x|} = |x| $\), we get:
   $$ |y| = e^C \cdot |x| $$

   Let \($ A = e^C $\), so we get:
   $$ y = A x $$

Thus, the solution to the fourth equation is:
$$ y = A x $$

## \($ \frac{dy}{dx} = x y $\)



1. **Rearrange the equation**:
   $$ \frac{dy}{dx} = x y \quad \Rightarrow \quad \frac{dy}{y} = x \, dx $$

2. **Integrate both sides**:
   $$ \int \frac{1}{y} \, dy = \int x \, dx $$

   This gives:
   $$ \ln |y| = \frac{x^2}{2} + C $$

3. **Exponentiate both sides** to solve for \($y$\):
   $$ |y| = e^{\frac{x^2}{2} + C} = e^C \cdot e^{\frac{x^2}{2}} $$

   Let \($ A = e^C $\), so we get:
   $$ y = A e^{\frac{x^2}{2}} $$

Thus, the solution to the fifth equation is:
$$ y = A e^{\frac{x^2}{2}} $$

# Second-order Ordinary Differential Equations

##Exercise 3: \($ y''(x) + y'(x) = 0 $\) with boundary conditions \($ y(0) = 2 $\) and \($ y'(0) = -1 $\)

We start with the second-order linear differential equation:

\[$
y''(x) + y'(x) = 0
$\]

### Formulate the characteristic equation

The equation is \($ y'' + y' = 0 $\), so the characteristic equation is:

\[$
r^2 + r = 0
$\]

### Solve the characteristic equation

Factoring the equation:

\[$
r(r + 1) = 0
$\]

So the roots are:

\[$
r = 0 \quad \text{and} \quad r = -1
$\]

### General solution

For the general solution, we use the fact that for roots \($r = 0$\) and \($r = -1$\), the solution is:

\[$
y(x) = C_1 + C_2 e^{-x}
$\]

### Apply boundary conditions

Now, we use the boundary conditions to find \($C_1$\) and \($C_2$\).

- **Boundary Condition 1: \($ y(0) = 2 $\)**

Substitute \($ x = 0 $\) into the general solution:

\[$
y(0) = C_1 + C_2 e^0 = C_1 + C_2 = 2
$\]

This gives the equation:

\[$
C_1 + C_2 = 2 \quad \text{(Equation 1)}
$\]

- **Boundary Condition 2: \($ y'(0) = -1 $\)**

Now, compute the derivative \($ y'(x) = -C_2 e^{-x} $\), and substitute \($ x = 0 $\):

\[$
y'(0) = -C_2 e^0 = -C_2 = -1
$\]

Thus, we have:

\[$
C_2 = 1
$\]

### Solve for \($C_1$\)

Substitute \($ C_2 = 1 $\) into Equation 1:

\[$
C_1 + 1 = 2 \quad \Rightarrow \quad C_1 = 1
$\]

Therefore, the solution is:

\[$
y(x) = 1 + e^{-x}
$\]

## \($ y''(x) - y(x) = 0 $\) with boundary conditions \($ y(0) = 2 $\) and \($ y'(0) = 0 $\)

We start with the second-order linear differential equation:

\[$
y''(x) - y(x) = 0
$\]

### Formulate the characteristic equation

The equation is \($ y'' - y = 0 $\), so the characteristic equation is:

\[$
r^2 - 1 = 0
$\]

### Solve the characteristic equation

Solving the characteristic equation:

\[$
r = \pm 1
$\]

### General solution

For the general solution, we use the fact that for the roots \($ r = 1 $\) and \($ r = -1 $\), the solution is:

\[$
y(x) = C_1 e^{x} + C_2 e^{-x}
$\]

### Apply boundary conditions

Now, we use the boundary conditions to find \($ C_1 $\) and \($ C_2 $\).

- **Boundary Condition 1: \($ y(0) = 2 \$)**

Substitute \($ x = 0 $\) into the general solution:

\[$
y(0) = C_1 e^0 + C_2 e^0 = C_1 + C_2 = 2
$\]

This gives the equation:

\[$
C_1 + C_2 = 2 \quad \text{(Equation 1)}
$\]

- **Boundary Condition 2: \($ y'(0) = 0 $\)**

Now, compute the derivative \($ y'(x) = C_1 e^x - C_2 e^{-x} $\), and substitute \($ x = 0 $\):

\[$
y'(0) = C_1 e^0 - C_2 e^0 = C_1 - C_2 = 0
$\]

Thus, we have:

\[$
C_1 = C_2
$\]

### Solve for \($ C_1 $\) and \($ C_2 $\)

Substitute \($ C_1 = C_2 $\) into Equation 1:

\[$
2C_1 = 2 \quad \Rightarrow \quad C_1 = 1
$\]

Thus, \($ C_1 = C_2 = 1 $\).

### Final Solution

The solution to the second equation is:

\[$
y(x) = e^x + e^{-x} = 2 \cosh(x)
$\]

# Wave Equation

Check if \($ \psi(t, x) = A \cos(\omega t + kx) $\) is a solution to the wave equation.

1. **Compute second partial derivatives**:
   - For time \( t \):
     $$ \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx) $$

   - For space \( x \):
     $$ \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx) $$

2. **Substitute into the wave equation**:
   The wave equation is:
   $$ \frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0 $$

   Substituting the derivatives:
   $$ -A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0 $$

   Factoring out common terms:
   $$ -A \cos(\omega t + kx) (\omega^2 - v^2 k^2) = 0 $$

3. **Conclusion**:
   For the equation to hold, we need:
   $$ \omega^2 = v^2 k^2 $$

   This condition is satisfied when \($ v = \frac{\omega}{k} $\), so the function \($ \psi(t, x) = A \cos(\omega t + kx) $\) is indeed a solution to the wave equation.
