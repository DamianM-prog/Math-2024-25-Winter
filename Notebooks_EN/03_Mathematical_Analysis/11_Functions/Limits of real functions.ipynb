# Universal Guide for Solving Limits and Asymptotes

## 1. **Finding Limits**

When solving limit problems, the method depends on the type of limit you're evaluating. Below are common types of limits with general steps for solving them.

### **Limit at Infinity:**

Given a rational function:

\[$
\lim_{x \to \infty} \frac{P(x)}{Q(x)}
$\]

Where \($P(x)$\) and \($Q(x)$\) are polynomials.

#### Steps to Solve:
- **Identify the highest degree terms** in both the numerator \($P(x)$\) and the denominator \($Q(x)$\).
- Divide both the numerator and the denominator by the highest power of \(x\) in the denominator.
- Simplify the expression, and evaluate the limit as \($x \to \infty$\).
- The limit will often be determined by the ratio of the highest degree terms in the numerator and denominator.

#### Special Cases:
- **Degree of numerator < degree of denominator**: The limit will be 0.
- **Degree of numerator = degree of denominator**: The limit is the ratio of the leading coefficients.
- **Degree of numerator > degree of denominator**: The limit is \($ \infty $\) or \($ -\infty $\), depending on the signs of the leading terms.

### **Limit at a Point (e.g., \($x \to a$\))**:

Given a function \(f(x)\), you can evaluate the limit by direct substitution:

\[$
\lim_{x \to a} f(x)
$\]

#### Steps to Solve:
- **Direct substitution**: Try to substitute \(x = a\) directly into the function. If the result is a defined number, that is the limit.
- If you encounter an indeterminate form like \($ \frac{0}{0} $\), use **algebraic manipulation** such as factoring, simplifying, or rationalizing the expression.
- **L'H么pital's Rule**: If you still get an indeterminate form like \($ \frac{0}{0} $\) or \($ \frac{\infty}{\infty} $\), apply L'H么pital's Rule. Differentiate the numerator and denominator separately, and then re-evaluate the limit.

#### Trigonometric Limits:
- For small angles, remember the standard limit:
  \[$
  \lim_{x \to 0} \frac{\sin(x)}{x} = 1
  $\]
- Similarly, \($\lim_{x \to 0} \frac{\tan(x)}{x} = 1$\).

---

## 2. **Finding Asymptotes**

Asymptotes are lines that a function approaches but never quite reaches. There are two main types: **vertical asymptotes** and **horizontal asymptotes**.

### **Vertical Asymptotes:**

A vertical asymptote occurs where the function goes to \($ \infty $\) or \($ -\infty $\) as \(x\) approaches a particular value.

Given a function:

\[$
f(x) = \frac{P(x)}{Q(x)}
$\]

#### Steps to Solve:
1. **Set the denominator equal to zero**: Solve \($Q(x) = 0$\). The values of \(x\) that make the denominator zero (but not the numerator) are potential vertical asymptotes.
2. **Check the behavior**: Ensure the numerator doesn't also become zero at the same point. If it does, the limit might be indeterminate and you should factor to simplify.
3. **Check the behavior as \(x\) approaches these points**: As \(x\) approaches the critical value from both sides, if \($f(x)$\) goes to \($ \infty $\) or \($ -\infty $\), then there is a vertical asymptote at that point.

### **Horizontal Asymptotes:**

Horizontal asymptotes describe the behavior of the function as \($x \to \infty$\) or \($x \to -\infty$\).

#### Steps to Solve:
- **For rational functions**:
  - If the degree of the numerator is less than the degree of the denominator, the horizontal asymptote is \(y = 0\).
  - If the degree of the numerator equals the degree of the denominator, the horizontal asymptote is \($y = \frac{a}{b}$\), where \(a\) and \(b\) are the leading coefficients of the numerator and denominator, respectively.
  - If the degree of the numerator is greater than the degree of the denominator, there is **no horizontal asymptote** (the function goes to \($ \infty $\) or \($ -\infty $\)).

- **For non-rational functions** (like trigonometric functions):
  - For functions like \($ \sin(x) $\), \($ \cos(x) $\), or other bounded functions in the numerator, the behavior as \($x \to \infty$\) or \($x \to -\infty$\) will often lead the function to approach 0 when the denominator grows without bound.
  
  Example:
  \[$
  \lim_{x \to \infty} \frac{\sin(x)}{x^2 + 1} = 0
  $\]
  because the denominator grows faster than the numerator oscillates.

---

## 3. **Summary of Key Concepts**

- **Limits at Infinity**: Focus on the highest degree terms in the numerator and denominator.
- **Limits at a Point**: Use direct substitution and algebraic simplification; apply L'H么pital's Rule if necessary.
- **Vertical Asymptotes**: Set the denominator equal to zero (but not the numerator), and analyze the function's behavior near those values.
- **Horizontal Asymptotes**: Analyze the degrees of the numerator and denominator for rational functions or evaluate the behavior at infinity for other functions.

---

## 4. **Common Tips & Tricks**

- For **rational functions**, when the numerator and denominator are both polynomials, focus on the degrees of the terms.
- When the limit leads to an **indeterminate form** like \($ \frac{0}{0} $\) or \($ \frac{\infty}{\infty} $\), simplify or apply L'H么pital's Rule.
- **Trigonometric limits** near 0: Use known approximations such as \($ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $\) and similar for \($\tan(x)$\).

### Exercise 1

We are given the limit:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$$

#### Step-by-step Solution:

1. **Rewrite the expression**:
   Divide both the numerator and denominator by \($x^4$\) to simplify the terms as \($ x \to \infty $\):

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = \lim_{x \to \infty} \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{\frac{x^4}{x^4} - \frac{3x^3}{x^4}}
$$

2. **Simplify the terms**:
   Now simplify each term:

$$
= \lim_{x \to \infty} \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}
$$

3. **Evaluate the limit**:
   As \($ x \to \infty $\), both \($ \frac{1}{x} $\) and \($ \frac{2}{x^2} $\) approach 0, and similarly \($ \frac{3}{x} $\) also approaches 0. So the expression becomes:

$$
= \frac{0 + 0}{1 - 0} = 0
$$

4. **Conclusion**:
   Thus, the limit is:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0
$$


### Exercise 2

We are given the limit:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}
$$

#### Step-by-step Solution:

1. **Apply a known limit**:
   We know the following standard limit:

$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

   We can use this to simplify \($ \sin(3x) $\). Notice that:

$$
\frac{\sin(3x)}{x} = 3 \cdot \frac{\sin(x)}{x}
$$

2. **Rewrite the limit**:
   Using the result from step 1, we can rewrite the original limit as:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = \lim_{x \to 0} \frac{3 \sin(x)}{2x + 1}
$$

3. **Evaluate the limit**:
   As \($ x \to 0 $\), we know that \($ \sin(x) \to 0 $\) and \($ 2x + 1 \to 1 $\). So the expression becomes:

$$
\frac{3(0)}{1} = 0
$$

4. **Conclusion**:
   Thus, the limit is:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = 0
$$


### Exercise 3

We are asked to find the asymptotes of the following functions:

#### a) For \($ f(x) = \frac{x^2 - 1}{x^2 + 1} $\):

1. **Rewrite the function**:
   
   $$ f(x) = \frac{x^2 - 1}{x^2 + 1} = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}} $$

2. **Evaluate the limit as \($ x \to \infty $\)**:
   
   As \($ x \to \infty $\), both \($ \frac{1}{x^2} $\) approach 0. So we have:

   $$ f(x) \to \frac{1 - 0}{1 + 0} = 1 $$

3. **Horizontal Asymptote**:
   
   The function approaches \($ y = 1 $\) as \($ x \to \infty $\), so the horizontal asymptote of \($ f(x) $\) is:

   $$ y = 1 $$

4. **Vertical Asymptotes**:
   
   There are no vertical asymptotes because the denominator \($ x^2 + 1 $\) is always positive and never equals zero.

#### b) For \($ g(x) = \frac{\sin(x)}{x^2 + 1} $\):

1. **Evaluate the limit as \($ x \to \infty $\)**:
   
   As \($ x \to \infty $\), the numerator \($ \sin(x) $\) remains bounded between -1 and 1, while the denominator \($ x^2 + 1 $\) grows without bound. Therefore, the function approaches 0 as \($ x \to \infty $\):

   $$ g(x) \to 0 $$

2. **Horizontal Asymptote**:
   
   The function approaches \($ y = 0 $\) as \($ x \to \infty $\), so the horizontal asymptote of \($ g(x) $\) is:

   $$ y = 0 $$

3. **Vertical Asymptotes**:
   
   There are no vertical asymptotes because the denominator \($ x^2 + 1 $\) is always positive and never zero.
