# Functions in Mathematics

A **function** is a relationship between two sets, where each input from the first set (domain) is associated with exactly one output in the second set (range). Functions are fundamental in mathematics and can be expressed in various forms.

## Definition of a Function

A function \( f \) maps an element \( x \) from the domain to an element \( y \) in the codomain:

$$
f: X \to Y
$$

### Function Notation

The notation \( f(x) \) represents the output of the function \( f \) at the input \( x \).

For example, for \($ f(x) = x^2 $\):

$$
f(3) = 3^2 = 9
$$

## Common Types of Functions

### 1. **Quadratic Function** \($ f(x) = x^2 $\)

The quadratic function is defined by squaring the input value \($ x $\). The graph of this function is a parabola.

$$
f(x) = x^2
$$

For example:

$$
f(2) = 2^2 = 4 \quad \text{and} \quad f(-3) = (-3)^2 = 9
$$

image.png
### 2. **Linear Function** \($ f(x) = x $\)

The linear function is a simple identity function where the output is equal to the input.

$$
f(x) = x
$$

For example:

$$
f(5) = 5 \quad \text{and} \quad f(-4) = -4
$$
image.png
### 3. **Rational Function** \($ f(x) = \frac{1}{x} $\)

The rational function \($ f(x) = \frac{1}{x} $\) is defined for all \($ x \neq 0 $\). Its graph has a vertical asymptote at \( x = 0 \).

$$
f(x) = \frac{1}{x}
$$

For example:

$$
f(2) = \frac{1}{2} \quad \text{and} \quad f(-1) = -1
$$

image.png
### 4. **Sine Function** \($ f(x) = \sin(x) $\)

The sine function is a periodic function that oscillates between -1 and 1. It is commonly used in trigonometry.

$$
f(x) = \sin(x)
$$

For example:

$$
f\left( \frac{\pi}{2} \right) = 1 \quad \text{and} \quad f\left( \pi \right) = 0
$$

image.png
### 5. **Exponential Function** \$( f(x) = e^x $\)

The exponential function is defined as \($ e^x $\), where \($ e \approx 2.718 $\) is the base of the natural logarithm. The function grows exponentially as \($ x $\) increases.

$$
f(x) = e^x
$$

For example:

$$
f(0) = e^0 = 1 \quad \text{and} \quad f(2) = e^2 \approx 7.389
$$
image.png
### 6. **Natural Logarithm** \($ f(x) = \ln(x) $\)

The natural logarithm function \($ \ln(x) $\) is the inverse of the exponential function. It is defined for \($ x > 0 $\).

$$
f(x) = \ln(x)
$$

For example:

$$
f(1) = \ln(1) = 0 \quad \text{and} \quad f(e) = \ln(e) = 1
$$

image.png
## Inverse Functions

An **inverse function** reverses the operation of the original function. If \($ f(x) $\) is a function, then the inverse function \($ f^{-1}(x) $\) satisfies:

$$
f(f^{-1}(x)) = x \quad \text{and} \quad f^{-1}(f(x)) = x
$$

### Example: Inverse of \($ f(x) = x^2 $\)

The inverse function of \($ f(x) = x^2 $\) is \($ f^{-1}(x) = \sqrt{x} $\), but it is defined only for \($ x \geq 0 $\) (the principal square root).

$$
f^{-1}(x) = \sqrt{x}
$$

For example:

$$
f^{-1}(9) = 3 \quad \text{and} \quad f^{-1}(16) = 4
$$

### Example: Inverse of \($ f(x) = \ln(x) $\)

The inverse of the natural logarithm function \($ \ln(x) $\) is the exponential function \($ e^x $\).

$$
f^{-1}(x) = e^x
$$

For example:

$$
f^{-1}(0) = e^0 = 1 \quad \text{and} \quad f^{-1}(1) = e^1 = e
$$
