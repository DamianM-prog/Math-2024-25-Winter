# Analytic Geometry: Conic Curves

## Introduction to Conic Curves
A conic curve (or simply conic) is a curve obtained by intersecting a plane with a double-napped cone. There are four types of conic curves:

1. **Circle**
2. **Ellipse**
3. **Parabola**
4. **Hyperbola**

Each of these curves can be represented by a second-degree equation in the form:

\[
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
\]

Where \( A \), \( B \), \( C \), \( D \), \( E \), and \( F \) are constants.

---

## Types of Conic Curves

### 1. Circle
A **circle** is the set of points in a plane that are equidistant from a fixed point called the **center**.

- **Equation of a Circle** (center \( (h, k) \) and radius \( r \)):

\[
(x - h)^2 + (y - k)^2 = r^2
\]

- **General Form**: \( x^2 + y^2 + Dx + Ey + F = 0 \)

### 2. Ellipse
An **ellipse** is the set of points where the sum of the distances from two fixed points (the foci) is constant.

- **Equation of an Ellipse** (center \( (h, k) \), semi-major axis \( a \), semi-minor axis \( b \)):


{(x - h)^2}{a^2} + {(y - k)^2}{b^2} = 1

- **General Form**: \( Ax^2 + By^2 + Dx + Ey + F = 0 \) where \( A \neq B \).

### 3. Parabola
### a. Vertex Form (Standard Form)
The vertex form of the equation of a parabola is:

\[
y = a(x - h)^2 + k
\]

Where:
- \( (h, k) \) is the **vertex** of the parabola.
- \( a \) controls the direction and width of the parabola:
  - If \( a > 0 \), the parabola opens **upwards**.
  - If \( a < 0 \), the parabola opens **downwards**.
  - Larger values of \( |a| \) make the parabola **narrower**, while smaller values make it **wider**.

### b. General Form (Standard Quadratic Form)
The general form of the equation of a parabola is:

\[
y = ax^2 + bx + c
\]

Where:
- \( a \), \( b \), and \( c \) are **constants**.
- This form does not explicitly show the vertex, but it can be converted to the vertex form by **completing the square** or using the **vertex formula**.

### c. Factored Form Using Roots \( x_1 \) and \( x_2 \)
If the parabola crosses the \( x \)-axis at two points \( x_1 \) and \( x_2 \), the equation can be written as:

\[
y = a(x - x_1)(x - x_2)
\]

Where:
- \( x_1 \) and \( x_2 \) are the **roots** (x-intercepts) of the parabola.
- \( a \) is a constant that determines the **width** and **direction** of the parabola:
  - If \( a > 0 \), the parabola opens **upwards**.
  - If \( a < 0 \), the parabola opens **downwards**.


- **General Form**: \( Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0 \) where \( A = C \).

### 4. Hyperbola
A **hyperbola** is the set of points where the absolute difference of the distances from two fixed points (the foci) is constant.

- **Equation of a Hyperbola** (center at \( (h, k) \), transverse axis length \( 2a \), conjugate axis length \( 2b \)):

\[
\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
\]

- **General Form**: \( Ax^2 + By^2 + Dx + Ey + F = 0 \) where \( A \neq B \) and the signs of \( A \) and \( B \) are opposite.

---

## Discriminant and Conic Classification
The discriminant \( \Delta \) of the general second-degree equation \( Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0 \) is used to classify the conic:

\[
\Delta = B^2 - 4AC
\]

- **If \( \Delta < 0 \)**: The conic is an **ellipse** or a **circle** (if \( A = C \)).
- **If \( \Delta = 0 \)**: The conic is a **parabola**.
- **If \( \Delta > 0 \)**: The conic is a **hyperbola**.

---

## Key Properties

- **Foci**: Points used to define ellipses, hyperbolas, and parabolas.
- **Directrix**: A line used to define parabolas.
- **Vertex**: A point on the conic that represents the closest point to the focus (in the case of a parabola).
- **Axes of Symmetry**: Lines that pass through the center of conics (major and minor axes for ellipses).


Exercise 1

A(1,2) r = 3

(x-1)^2 + (y-2) = 9

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbols for x and y
x, y = sp.symbols('x y')

# Define the equation of the circle (x-1)^2 + (y-2)^2 = 9
circle_eq = (x - 1)**2 + (y - 2)**2 - 9

# To check the equation
print("Equation of the circle:", circle_eq)

# Plot the circle
# Convert the equation into a form suitable for plotting
# We know that the equation is (x-1)^2 + (y-2)^2 = 9
# So, the center is at (1, 2) and the radius is 3 (since sqrt(9) = 3)

center = (1, 2)
radius = 3

# Generate values for the circle using parametric equations
theta = np.linspace(0, 2 * np.pi, 400)
x_vals = center[0] + radius * np.cos(theta)
y_vals = center[1] + radius * np.sin(theta)

# Plot the circle
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals, label=r'$(x-1)^2 + (y-2)^2 = 9$', color='b')
plt.scatter(center[0], center[1], color='red', label='Center (1, 2)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')

# Adding labels and title
plt.title('Graph of the Circle $(x-1)^2 + (y-2)^2 = 9$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

1 = -a

a = -1

y = -(x-2)(x-4)

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable
x = sp.symbols('x')

# Define the equation of the parabola with unknown 'a'
a = sp.symbols('a')
parabola_eq = a * (x - 2) * (x - 4)

# Solve for 'a' using the point (3, 1)
a_value = sp.solve(parabola_eq.subs(x, 3) - 1, a)[0]

# Now we have the equation of the parabola with the correct 'a'
parabola = parabola_eq.subs(a, a_value)

# Convert the equation to a function for plotting
parabola_func = sp.lambdify(x, parabola, "numpy")

# Define the range for plotting
x_vals = np.linspace(0, 6, 400)
y_vals = parabola_func(x_vals)

# Plot the parabola
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Parabola: $y = a(x - 2)(x - 4)$', color='b')

# Plot the point (3, 1)
plt.scatter(3, 1, color='red', label='Point (3, 1)', zorder=5)

# Highlight the x-intercepts
plt.axhline(0, color='black',linewidth=1)
plt.axvline(2, color='green', linestyle='--', label='x = 2')
plt.axvline(4, color='green', linestyle='--', label='x = 4')

# Labels and title
plt.title('Graph of the Parabola')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show grid and plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values for plotting
x = np.linspace(-3, 3, 400)

# Hyperbola equation: x^2 - y^2 = 1  --> y = ± sqrt(x^2 - 1)
y_hyperbola_positive = np.sqrt(x**2 - 1)
y_hyperbola_negative = -np.sqrt(x**2 - 1)

# Ellipse equation: x^2 + 4y^2 = 6 --> y = ± sqrt((6 - x^2)/4)
y_ellipse_positive = np.sqrt((6 - x**2) / 4)
y_ellipse_negative = -np.sqrt((6 - x**2) / 4)

# Plot the hyperbola
plt.figure(figsize=(6, 6))
plt.plot(x, y_hyperbola_positive, label=r'$x^2 - y^2 = 1$ (Hyperbola)', color='blue')
plt.plot(x, y_hyperbola_negative, color='blue')

# Plot the ellipse
plt.plot(x, y_ellipse_positive, label=r'$x^2 + 4y^2 = 6$ (Ellipse)', color='green')
plt.plot(x, y_ellipse_negative, color='green')

# Mark the intersection points
# We already know the intersection points: (sqrt(2), 1), (-sqrt(2), 1), (sqrt(2), -1), (-sqrt(2), -1)
intersection_x = [np.sqrt(2), -np.sqrt(2), np.sqrt(2), -np.sqrt(2)]
intersection_y = [1, 1, -1, -1]

# Plot the intersection points
plt.scatter(intersection_x, intersection_y, color='red', label='Intersection Points')

# Labels and title
plt.title("Hyperbola and Ellipse Intersection")
plt.xlabel('x')
plt.ylabel('y')

# Set the aspect ratio to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Add a legend
plt.legend()

# Display the grid
plt.grid(True)

# Show the plot
plt.show()
```

x^2 - y^2 = 1

x^2 = y^2 +1

D = I sqrt/y^2+1 - (-sqrt/y^2+1)I

D = 2sqrt/y^2+1

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the y values for plotting (range of y-values)
y = np.linspace(-3, 3, 400)

# Calculate the corresponding x-values for the hyperbola branches: x = ± sqrt(y^2 + 1)
x_positive = np.sqrt(y**2 + 1)
x_negative = -np.sqrt(y**2 + 1)

# Calculate the distance between the branches at each y value
D = 2 * np.sqrt(y**2 + 1)

# Plot the hyperbola
plt.figure(figsize=(8, 6))

# Plot the positive and negative branches of the hyperbola
plt.plot(x_positive, y, label=r'$x^2 - y^2 = 1$ (Hyperbola - Positive Branch)', color='blue')
plt.plot(x_negative, y, label=r'$x^2 - y^2 = 1$ (Hyperbola - Negative Branch)', color='blue')

# Plot the distance between the branches (as a function of y)
plt.plot(D, y, label=r'Distance Between Branches: $D(y) = 2 \sqrt{y^2 + 1}$', color='red', linestyle='--')

# Labels and title
plt.title("Hyperbola and Distance Between Branches")
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Display the grid
plt.grid(True)

# Show the plot
plt.show()
```
