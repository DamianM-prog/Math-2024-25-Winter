### Exercise 3
\[$
f'(x) = \frac{d}{dx} \left( x^2 + 3x - 5 \right)
$\]
\[$
f'(x) = 2x + 3
$\]

\[$
2x + 3 = 0
$\]

Solving for \( x \):

\[$
2x = -3
$\]
\[$
x = -\frac{3}{2}
$\]

\($ x = -\frac{3}{2} $\) is a critical point.

\[$
f''(x) = \frac{d}{dx} (2x + 3)
$\]
\[$
f''(x) = 2
$\]

Since the second derivative \($ f''(x) = 2 $\) is positive, the function is concave up at \($ x = -\frac{3}{2} $\), indicating that the critical point is a **minimum**.

\[$
f\left( -\frac{3}{2} \right) = \left( -\frac{3}{2} \right)^2 + 3\left( -\frac{3}{2} \right) - 5
$\]
\[$
f\left( -\frac{3}{2} \right) = \frac{9}{4} - \frac{9}{2} - 5
$\]

Convert all terms to have a denominator of 4:

\[$
f\left( -\frac{3}{2} \right) = \frac{9}{4} - \frac{18}{4} - \frac{20}{4}
$\]
\[$
f\left( -\frac{3}{2} \right) = \frac{9 - 18 - 20}{4} = \frac{-29}{4}
$\]

\[$
f\left( -\frac{3}{2} \right) = -\frac{29}{4}
$\]

- The critical point is \($ x = -\frac{3}{2} $\).
- This is a **minimum** point.
- The minimum value of the function is \($ f\left( -\frac{3}{2} \right) = -\frac{29}{4} $\).


### Exercise 4

\[$
f(x) = \frac{x^2 + 2x + 1}{x - 1}
$\]

\[$
f(x) = \frac{(x + 1)^2}{x - 1}
$\]

\($ f(x) = \frac{g(x)}{h(x)} $\)

\[$
f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{[h(x)]^2}
$\]

 \($ g(x) = (x + 1)^2 $\)

 \($ h(x) = x - 1 $\)

First, compute the derivatives of \( g(x) \) and \( h(x) \):
- \($ g'(x) = 2(x + 1) $\)
- \($ h'(x) = 1 $\)

\[$
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2 \cdot 1}{(x - 1)^2}
$\]

\[$
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2}{(x - 1)^2}
$\]
\[$
= \frac{2(x^2 - 1) - (x^2 + 2x + 1)}{(x - 1)^2}
$\]
\[$
= \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2}
$\]
\[$
= \frac{x^2 - 2x - 3}{(x - 1)^2}
$\]


\[$
\frac{x^2 - 2x - 3}{(x - 1)^2} = 0
$\]

The numerator must be equal to zero:

\[$
x^2 - 2x - 3 = 0
$\]


\[$
x^2 - 2x - 3 = (x - 3)(x + 1) = 0
$\]

 \( x = 3 \) or \( x = -1 \).


1. **At \( x = 3 \)**:
    - Choose test points to the left and right of \( x = 3 \). \( x = 2 \) and \( x = 4 \).
    - For \( x = 2 \), into \( f'(x) \):
    
      \[$
      f'(2) = \frac{2^2 - 2(2) - 3}{(2 - 1)^2} = \frac{4 - 4 - 3}{1} = -3 \quad (\text{negative})
      $\]

    - For \( x = 4 \),  into \( f'(x) \):

      \[$
      f'(4) = \frac{4^2 - 2(4) - 3}{(4 - 1)^2} = \frac{16 - 8 - 3}{9} = \frac{5}{9} \quad (\text{positive})
      $\]

    - Since \( f'(x) \) changes from negative to positive as \( x \) increases through \( x = 3 \), \( x = 3 \) is a **local minimum**.

2. **At \( x = -1 \)**:
    - Choose test points to the left and right of \( x = -1 \). \( x = -2 \) and \( x = 0 \).
    - For \( x = -2 \),  into \( f'(x) \):

      \[$
      f'(-2) = \frac{(-2)^2 - 2(-2) - 3}{(-2 - 1)^2} = \frac{4 + 4 - 3}{9} = \frac{5}{9} \quad (\text{positive})
      $\]

    - For \( x = 0 \),  into \( f'(x) \):

      \[$
      f'(0) = \frac{0^2 - 2(0) - 3}{(0 - 1)^2} = \frac{-3}{1} = -3 \quad (\text{negative})
      $\]

    - Since \( f'(x) \) changes from positive to negative as \( x \) decreases through \( x = -1 \), \( x = -1 \) is a **local maximum**.

- The critical points are \( x = 3 \) and \( x = -1 \).
