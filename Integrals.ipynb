# Essential Information About Integrals

## What is an Integral?

An **integral** is a mathematical operation used to calculate areas under curves, volumes, and other quantities that accumulate or vary continuously. In simple terms, it represents the accumulation of a quantity over a given interval. There are two main types of integrals:

1. **Definite Integral**: Computes the area under a curve within a specified interval.
2. **Indefinite Integral**: Represents a general function whose derivative is the integrand.

### Definite Integral

A **definite integral** computes the total accumulation of a function over a specific interval. It is written as:

$$
\int_a^b f(x) \, dx
$$

where \($ f(x) $\) is the function being integrated, and \($ [a, b] $\) is the interval over which we are integrating. The result of a definite integral is a numerical value, representing the net area under the curve between \($ x = a $\) and \($ x = b $\).

For example, the integral of \($ f(x) = 2x + 1 $\) from 0 to \($ \pi $\) is:

$$
\int_0^\pi (2x + 1) \, dx = \left[ x^2 + x \right]_0^\pi = \pi^2 + \pi
$$

### Indefinite Integral

An **indefinite integral** represents a family of functions, as it does not have specific bounds. It is written as:

$$
\int f(x) \, dx = F(x) + C
$$

where \($ F(x) $\) is the antiderivative of \($ f(x) $\), and \($ C $\) is the constant of integration. The result of an indefinite integral is a general expression, representing all possible functions whose derivative is the integrand.

For example, the indefinite integral of \($ f(x) = 3x^2 $\) is:

$$
\int 3x^2 \, dx = x^3 + C
$$

### Properties of Definite Integrals

1. **Linearity of Integration**: The integral of a sum of functions is the sum of the integrals.

$$
\int_a^b \left( f(x) + g(x) \right) \, dx = \int_a^b f(x) \, dx + \int_a^b g(x) \, dx
$$

2. **Constant Multiple Rule**: A constant can be factored out of the integral.

$$
\int_a^b c \cdot f(x) \, dx = c \cdot \int_a^b f(x) \, dx
$$

3. **Reversal of Limits**: Reversing the limits of integration changes the sign of the integral.

$$
\int_a^b f(x) \, dx = -\int_b^a f(x) \, dx
$$

### Fundamental Theorem of Calculus

The **Fundamental Theorem of Calculus** connects the concept of differentiation with integration. It consists of two parts:

1. **First Part**: If \($ F(x) $\) is the antiderivative of \($ f(x) $\), then:

$$
\int_a^b f(x) \, dx = F(b) - F(a)
$$

2. **Second Part**: If \($ f(x) $\) is continuous on an interval \($ [a, b] $\), then the function \($ F(x) $\) defined by:

$$
F(x) = \int_a^x f(t) \, dt
$$

is the antiderivative of \($ f(x) $\).

### Techniques of Integration

Several methods can be used to compute integrals, depending on the form of the integrand:

1. **Substitution**: Useful when the integrand is a composition of functions. The idea is to simplify the integral by making a substitution that reduces the complexity of the integral.

2. **Integration by Parts**: Based on the product rule for derivatives, this technique is useful when integrating products of functions.

$$
\int u \, dv = uv - \int v \, du
$$

3. **Partial Fractions**: Decomposes a rational function into simpler fractions that are easier to integrate.

4. **Trigonometric Integrals**: Specific techniques for integrating functions involving trigonometric functions, such as using trigonometric identities.

### Area Under a Curve

The area under the curve of a function \($ f(x) $\) between two points \($ a $\) and \($ b $\) is given by the definite integral:

$$
A = \int_a^b f(x) \, dx
$$

This represents the total accumulated area between the graph of \($ f(x) $\) and the x-axis over the interval \($ [a, b] $\).

### Length of a Curve

The length \($ L $\) of a curve \($ y = f(x) $\) from \($ x = a $\) to \($ x = b $\) is given by:

$$
L = \int_a^b \sqrt{1 + \left( f'(x) \right)^2} \, dx
$$

### Example of a Definite Integral

For the function \($ f(x) = x^2 + 1 $\), to calculate the area under the curve from \($ x = 1 $\) to \($ x = 2 $\), we use the integral:

$$
A = \int_1^2 (x^2 + 1) \, dx = \left[ \frac{x^3}{3} + x \right]_1^2 = \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right) = \frac{7}{3}
$$

This result represents the total area under the curve from \($ x = 1 $\) to \($ x = 2 $\).

## Find the distance of the moving particle between time $t = 0$ and $t = 2$ for the position function $x(t) = 3t^2 - 6t + 1$.

The velocity function is the derivative of the position function:

$$
v(t) = \frac{d}{dt}(3t^2 - 6t + 1) = 6t - 6
$$

The distance traveled is the integral of the absolute value of the velocity function:

$$
\text{Distance} = \int_0^2 |6t - 6| \, dt
$$

For $t = 0$ to $t = 1$, $6t - 6$ is negative, and for $t = 1$ to $t = 2$, it is positive. Therefore, we compute:

$$
\text{Distance} = \int_0^1 -(6t - 6) \, dt + \int_1^2 (6t - 6) \, dt
$$

First integral:

$$
\int_0^1 -(6t - 6) \, dt = \left[ -3t^2 + 6t \right]_0^1 = -3 + 6 = 3
$$

Second integral:

$$
\int_1^2 (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_1^2 = 12 - 12 - (3 - 6) = 3
$$

Thus, the total distance is:

$$
\text{Distance} = 3 + 3 = 6
$$
